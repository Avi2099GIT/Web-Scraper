```python
from bs4 import BeautifulSoup
import csv
import requests

#Inorder to scrape the given list of websites we use the Beautiful SOup Library. It is mainly used to extract data from HTML and XML documents/ webpages.
link_file = 'WebLinks.csv'
with open(link_file) as csv_file:
    file_reader = csv.reader(csv_file)
    
    for row in file_reader:

        link = row[0]
        URL = requests.get(link)
        soup = BeautifulSoup(URL.text, 'lxml')
        article=soup.find('article')
        
        headline = article.div.h1.text
        text = soup.get_text()
        
        print("==========================================================================================================")
        print(link)
        print("==========================================================================================================")
        print("The headline of the given link: ", headline)
        print("==========================================================================================================")
        print("The text enclosed in the link: ", text)
        print("==========================================================================================================")
```
